```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

faq_questions = [
    "What is Python?",
    "What is AI?",
    "What is Machine Learning?",
    "What is Deep Learning?",
    "What is Data Science?",
    "What is NLP?",
    "What is ChatGPT?",
    "What is a chatbot?",
    "What is programming?",
    "What is coding?",
    "What is GitHub?",
    "What is an API?",
    "What is cloud computing?",
    "What is cybersecurity?",
    "What is database?",
    "What is computer science?",
    "What is software?"
]

faq_answers = [
    "Python is a programming language.",
    "AI stands for Artificial Intelligence.",
    "Machine Learning is a branch of AI.",
    "Deep Learning is a subset of Machine Learning.",
    "Data Science is the study of data.",
    "NLP stands for Natural Language Processing.",
    "ChatGPT is an AI chatbot developed by OpenAI.",
    "A chatbot is a program that interacts with users.",
    "Programming is the process of creating software.",
    "Coding means writing instructions for computers.",
    "GitHub is a platform for hosting code.",
    "An API allows applications to communicate.",
    "Cloud computing provides services over the internet.",
    "Cybersecurity protects systems from cyber attacks.",
    "A database stores and manages data.",
    "Computer Science is the study of computers and algorithms.",
    "Software is a set of instructions that tells a computer what to do."
]

vectorizer = TfidfVectorizer()
faq_vectors = vectorizer.fit_transform(faq_questions)

print("FAQ Chatbot Started!")
print("Type 'exit' to stop.\n")

while True:
    user_question = input("Shree: ")

    if user_question.lower() == "exit":
        print("nexa: Goodbye!")
        break

    user_vector = vectorizer.transform([user_question])
    similarity = cosine_similarity(user_vector, faq_vectors)

    best_match_index = similarity.argmax()
    best_score = similarity[0][best_match_index]

    if best_score < 0.6:
        print("nexa: Sorry, I don't know the answer.")
    else:
        print("nexa:", faq_answers[best_match_index])
```
